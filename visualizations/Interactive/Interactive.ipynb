```python
import pandas as pd
from datetime import datetime
```

```python
crime = pd.read_csv('crime.csv')
```

```python
crime['REPORTED_DATE'] = crime['REPORTED_DATE'].apply(lambda x: datetime.strptime(x.split(' ')[0], '%m/%d/%Y'))
```

```python
crime.head()
```

```python
date_count = crime[crime['IS_CRIME'] == 1].groupby(['REPORTED_DATE'])['INCIDENT_ID'].count().reset_index()
```

```python
date_count.head()
```

```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
trace = go.Scatter(x=date_count['REPORTED_DATE'],
                   y=date_count['INCIDENT_ID'])

data = [trace]
layout = dict(
    title='Crime Over Time',
    xaxis=dict(
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
plotly.offline.plot(fig, validate=False, filename='InteractiveTimeSeries.html')
```
