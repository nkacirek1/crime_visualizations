```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crime = pd.read_csv('crime.csv')
```

```python
crime.columns
```

```python
crime_by_pre = crime.groupby(['DISTRICT_ID','PRECINCT_ID'])['IS_CRIME'].sum().reset_index()
```

```python
crime_by_pre.columns = ['District ID', 'Precinct ID', 'Number of Crimes']
```

```python
plt.style.use('ggplot')
```

```python
sns.lmplot(x="Precinct ID", y="Number of Crimes", data=crime_by_pre, fit_reg=False, hue='District ID', 
           legend=True, height=6, aspect=2, scatter_kws={"s": 100})
plt.title('Number Crimes Reported by Precinct ID\n')
plt.xlabel('\nPrecinct ID\n')
plt.ylabel('Number of Crimes\n')
plt.savefig('Scatterplot.png')
```
