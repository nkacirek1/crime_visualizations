```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
shoot = pd.read_csv('denver_police_officer_involved_shootings.csv')
```

```python
shoot.head()
```

```python
plt.hist('AGE', data=shoot[shoot['ROLE']=='Officer'], alpha=0.6)
plt.title('Age of Officer Involved in Shooting\n')
plt.xlabel('\nAge')
plt.ylabel('Number of Officers\n')
plt.tight_layout()
plt.savefig('Histogram.png')
```
