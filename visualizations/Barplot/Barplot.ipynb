```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
crime = pd.read_csv('crime.csv')
```

```python
# get the count of each of the different offense categories 
for_bar = np.unique(crime.OFFENSE_CATEGORY_ID.values, return_counts=True)
```

```python
# group for sorting
id_count = list(zip(for_bar[0], for_bar[1]))
id_count.sort(key = lambda x: x[1], reverse=True)
```

```python
id_count
```

```python
seperate = [list(t) for t in zip(*id_count)]
```

```python
plt.barh(seperate[0], seperate[1])
plt.xlabel('\nCount')
plt.ylabel('Offense Category\n')
plt.title('Number of Crimes by Offense Category\n')
plt.xticks(np.arange(0, 124387, step=10000))
plt.tight_layout() # otherwise cut off the y axis label
plt.savefig('Barplot.png')
```
